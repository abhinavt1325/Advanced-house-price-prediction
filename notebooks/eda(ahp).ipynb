```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.head()
```

```python
dataset.shape
```

```python
# Calculate and display the percentage of missing values for each feature in the dataset
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean(),4))
```

```python
for feature in features_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
numerical_features=[feature for feature in dataset.columns if dataset[feature].dtypes!='O']
print("No. of numerical features:", len(numerical_features))
dataset[numerical_features].head()
```

```python
year_feature=[feature for feature in numerical_features if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
for feature in year_feature:
    print(feature,dataset[feature].unique())
```

```python
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title('Year Sold vs Median Price')
```

```python
for feature in year_feature:
    if feature!='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
discrete_feature=[feature for feature in numerical_features 
                  if len(dataset[feature].unique())<25 and feature not in year_feature+['Id']]
print('No. of discrete features', len(discrete_feature))
discrete_feature
```

```python
for feature in discrete_feature:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
continuous_feature=[feature for feature in numerical_features 
                  if feature not in discrete_feature+year_feature+['Id']]
print('No. of continuous features', len(continuous_feature))
continuous_feature
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```

```python
categorical_features=[feature for feature in dataset.columns if dataset[feature].dtypes == 'O']
categorical_features
```

```python
dataset[categorical_features].head()
```

```python
for feature in categorical_features:
    print("The feature {} has {} no. of categories".format(feature,len(dataset[feature].unique())))
```

```python
for feature in categorical_features:
    data=dataset.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```
